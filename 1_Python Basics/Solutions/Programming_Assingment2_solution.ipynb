1.	Write a Python program to convert kilometers to miles?

```python
kilometers = float(input("Enter the value in Kilometer"))

# 1 kilometer is equal to 0.621371 miles
kilo = 0.621371

# calculate miles
miles = kilometers * kilo
print(f'{kilometers} is equal to {round(miles,2)}')
```

2.	Write a Python program to convert Celsius to Fahrenheit?

```python

celsius = float(input("Enter the temprature in celcius "))

# calculate fahrenheit
fahrenheit = (celsius * 1.8) + 32
print(f'{celsius} is equal to {round(fahrenheit,2)} degree Fahrenheit')

# print('%0.1f degree Celsius is equal to %0.1f degree Fahrenheit' %(celsius,fahrenheit))
```

4.	Write a Python program to solve quadratic equation?

```python
import cmath
```

```python
a = 1
b = 5
c = 6

# calculate the discriminant
d = (b**2) - (4*a*c)

# find two solutions
sol1 = (-b-cmath.sqrt(d))/(2*a)
sol2 = (-b+cmath.sqrt(d))/(2*a)

print(f'The solution are {sol1} and {sol2}')
```
